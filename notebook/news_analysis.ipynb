```python
import matplotlib.pyplot as plt

from src.utils import find_top_websites
from src.loader import NewsDataLoader

# Initialize the data loader
data_loader = NewsDataLoader()

# Load the news data
news_dataset_path = '../data/data.csv'
news_data = data_loader.load_data(news_dataset_path)

# Find the top websites
top_news_websites = find_top_websites(news_data)
print(top_news_websites)

# Plot the data
plt.figure(figsize=(10,6))
top_news_websites.plot(kind='bar',color='blue')
plt.title('Top websites with the largest count of news articles')
plt.xlabel('Website Domain')
plt.ylabel('No of News Articles')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
import sys
import os
import matplotlib.pyplot as plt

# Ensure the parent directory is in the system path
parent_dir = os.path.abspath("..")
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

from src.utils import find_high_traffic_websites
from src.loader import NewsDataLoader

# Initialize the data loader
data_loader = NewsDataLoader()

# Load the traffic data
traffic_dataset_path = '../data/traffic.csv'
traffic_data = data_loader.load_data(traffic_dataset_path)

# Find the top traffic websites
top_traffic_websites = find_high_traffic_websites(traffic_data)
print(top_traffic_websites)

# Plot the data
plt.figure(figsize=(10,6))
top_traffic_websites.plot(kind='line',marker='o',color='green',linestyle='-')
plt.title('Top websites with the highest traffic')
plt.xlabel('Website')
plt.ylabel('Total IPs')
plt.xticks(range(len(top_traffic_websites.index)), top_traffic_websites.index, rotation=55)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import sys
import os
import matplotlib.pyplot as plt
from src.utils import find_countries_with_most_media
from src.loader import NewsDataLoader

def plot_countries_with_most_media(file_path):
    """
    Load the domain location data, find the countries with the most media organizations, and plot the data.
    """
    # Ensure the parent directory is in the system path
    parent_dir = os.path.abspath("..")
    if parent_dir not in sys.path:
        sys.path.insert(0, parent_dir)

    # Initialize the data loader
    data_loader = NewsDataLoader()

    # Load the domain location data
    domain_location_data = data_loader.load_data(file_path)

    # Check if data is loaded successfully
    if domain_location_data is None:
        print("Failed to load data.")
        return

    # Find the countries with the most media organizations
    countries_with_most_media = find_countries_with_most_media(domain_location_data)
    print(countries_with_most_media)

    # Plot the data
    plt.figure(figsize=(8,8))
    plt.pie(countries_with_most_media, labels=countries_with_most_media.index, autopct='%1.1f%%')
    plt.title('Countries with the highest number of news media organizations')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

# Use the function
domain_location_dataset_path = '../data/domains_location.csv'
plot_countries_with_most_media(domain_location_dataset_path)
```

```python
import sys, os
if os.path.abspath("..") not in sys.path:
    sys.path.insert(0, os.path.abspath(".."))

#a visual plot library
import matplotlib.pyplot as plt
from src.loader import NewsDataLoader
from src.utils import find_popular_articles


loader = NewsDataLoader()
popular_countries_data = loader.load_data('../data/rating.csv')
popular_articles = find_popular_articles(popular_countries_data)

countries = [country for (country, _), count in popular_articles]
counts = [count for (_, _), count in popular_articles]

plt.figure(figsize=(10, 6))
plt.bar(countries, counts, color='red')
plt.title('Top 10 Countries Mentioned in Articles')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import sys, os
if os.path.abspath("..") not in sys.path:
    sys.path.insert(0, os.path.abspath(".."))

import matplotlib.pyplot as plt

from src.utils import website_sentiment_distribution
from src.loader import NewsDataLoader

loader = NewsDataLoader()
sentiment_data = loader.load_data('../data/rating.csv')


sentiment_counts = website_sentiment_distribution(sentiment_data)


top_domains = sentiment_counts.sort_values(by='Total',ascending=False).head(10)

# Plot the mean sentiment counts of the top domains
top_domains[['Mean']].plot(kind='bar',figsize=(10,6))
plt.title('Mean Sentiment Counts of Top 10 Domains')
plt.xlabel('Domain')
plt.ylabel('Mean Sentiment Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Plot the median sentiment counts of the top domains
top_domains[['Median']].plot(kind='bar', figsize=(10, 6))
plt.title('Median Sentiment Counts of Top 10 Domains')
plt.xlabel('Domain')
plt.ylabel('Median Sentiment Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
